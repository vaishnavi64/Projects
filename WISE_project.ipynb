```python
data = data.drop(columns = ['YELLOW_FINGERS','PEER_PRESSURE','CHRONIC DISEASE','ALCOHOL CONSUMING','ANXIETY'])
data
```

```python
data = data.drop(columns = ['ALLERGY '])
data
```

```python
data = pd.get_dummies(data)
data
```

```python
data = data.drop(columns = ['GENDER_F','LUNG_CANCER_NO'])
data
```

```python
target = data['LUNG_CANCER_YES']

predict = data.drop(columns = ['LUNG_CANCER_YES'])
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
predict = ss.fit_transform(predict)
```

```python
predict[5]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(predict, target, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
lr_predicted = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(lr_predicted, y_test))
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear', random_state = 0)
svm.fit(x_train, y_train)
```

```python
y_pred = svm.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive_bayes = GaussianNB()
naive_bayes.fit(x_train, y_train)
```

```python
y_pred = naive_bayes.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
criterion = 'gini'
decision_tree = DecisionTreeClassifier(criterion=criterion)
decision_tree.fit(x_train, y_train)
```

```python
y_pred = decision_tree.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
criterion = 'gini'
random_forest = RandomForestClassifier(criterion=criterion)
random_forest.fit(x_train, y_train)
```

```python
y_pred = random_forest.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
